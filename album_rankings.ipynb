```python
def get_album_rankings(url):
    row = "o-chart-results-list-row-container"
    t_row = "div"

    artist_name = "c-title a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only"
    top_artist_name = "c-title a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 u-font-size-23@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-245 u-max-width-230@tablet-only u-letter-spacing-0028@tablet"

    rank = "c-label  a-font-primary-bold-l u-font-size-32@tablet u-letter-spacing-0080@tablet"
    t_rank = "span"

    next_album = "c-title a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only"
    top_album = "c-title a-font-primary-bold-l a-font-primary-bold-m@mobile-max lrv-u-color-black u-color-white@mobile-max lrv-u-margin-r-150"
    t_album = "h3"

    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    date = pd.to_datetime(soup.find_all(id="section-heading")[1].text.strip().strip("Week of "))

    if soup.text.find("Taylor Swift") < 0:
        print("null")

    rows = soup.find_all(t_row, row)

    rank = np.nan
    album_ranking = []

    for row in rows:
        if row.text.find("Taylor Swift") > 0:
            album = row.find("h3").text.strip()
            rank = row.find("span").text.strip()

            album_ranking.append([date, album, rank])
            #album_rankings.append(album_ranking)

    return(album_ranking)
```

```python
df = []
start = datetime.now()
base_url = "https://www.billboard.com/charts/billboard-200/"
b = pd.date_range(start = pd.to_datetime('2006-10-31'), end=datetime.now(), freq="W-Tue")
dates = [str(date).split()[0] for date in b]

for date in dates:
    url = base_url + date
    rankings = get_album_rankings(url)
    if df is not None:
        df.extend(rankings)
    print(date)
print(datetime.now() - start)

a = pd.DataFrame(df, columns = ['date', 'album', 'rank'])
a['rank'] = pd.to_numeric(a['rank'])
a = a.pivot(index = 'date', columns = 'album', values = 'rank')
a.to_csv('../data/album_rankings.csv', sep='\t')
```
